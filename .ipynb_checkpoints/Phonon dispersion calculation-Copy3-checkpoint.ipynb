```python
def plot_oscillation(y_ls):
    num_pos = len(y_ls)//10
    split = []
    temp = []

    for i in range(len(y_ls)):

        if i != 0 and i % 10 == 0:

            split.append(np.array(temp))
            temp = []

        temp.append(y_ls[i])
    split.append(np.array(temp))

    for i in range(len(split)):
        split[i] = (split[i]-np.average(split[i]))/np.max(abs(split[i]-np.average(split[i])))
    res = []
    
    
    
    for i in range(len(split)):
        plt.plot(split[i],'o-')
    

def phase_shift(y_ls):
    num_pos = len(y_ls)//10
    split = []
    temp = []

    for i in range(len(y_ls)):

        if i != 0 and i % 10 == 0:

            split.append(np.array(temp))
            temp = []

        temp.append(y_ls[i])
    split.append(np.array(temp))

    for i in range(len(split)):
        split[i] = (split[i]-np.average(split[i]))/np.max(abs(split[i]-np.average(split[i])))
    res = []

    for i in range(len(split)):
        a = np.fft.fft(split[i])
        res.append(np.mod(np.angle(a[1]),2*np.pi))
    return np.array(res)
```

```python
def initialize(M, m_ls, K1, K2, K_intra, T_amp, w0, sigma):
    M_intra = Matrix_intra(M,m_ls,K_intra)

    return ([M_intra, M, K1, K2, K_intra, m_ls],[T_amp, w0, sigma])

def energy_den_SingleQ(x, t, qx, qy, sys_par, dr_par):

    w_ls, eigv_ls = Eig(qx,qy,sys_par)

    velocities = [phonon_v(eigv_ls[i],w_ls[i],qx,qy,sys_par) for i in range(10)]
    res = 0
    
    for v in velocities:
        if (np.linalg.norm(v) > 0.0000000001):
            res += energy_den_SingleV_rescaled_by_V(x, t, dr_par, v)

        
    return res

def energy_den_SingleV_rescaled_by_V(x, t, dr_par, ph_v):

    T_amp, w0, sigma = dr_par

    vx = ph_v[0]
    v = np.linalg.norm(ph_v)
    
    
    x0 = x * vx / v
    
    if (x0**2 > x**2):
        y0 = 0
    else:
        y0 = np.sqrt(x**2-x0**2)
    
    prefactor =  (1/(2*np.pi*sigma**2)) * np.exp(-y0**2/(2*sigma**2)) * (T_amp)
    

    I = (-sigma/(4*v))*np.sqrt(8) * (2*np.real(np.exp(-1j * w0 * (t - x0/v)) * \
        np.exp(-(x0**2*v+2j*w0*x0*sigma**2)/(2*v*sigma**2)) * \
        dawsn((-2j*x0*v + 2*sigma**2*w0)/(v* sigma * np.sqrt(8)))))
    I_inf = -np.sqrt(np.pi / 2) * sigma / v *np.exp(-0.5 * (sigma * w0 / v)**2)*np.sin((t - x0/v)*w0)

    # add 1 because of the negative 1 at negative infinity
    I_from_1 = (sigma*np.sqrt(np.pi/2)/v) * (erf(x0/(sigma*np.sqrt(2))) + 1)   

    
    return prefactor * ((I-I_inf) + I_from_1) * v



def phonon_v(vec,w,qx,qy,sys_par):
    
    M  = sys_par[1]
    K1 = sys_par[2]
    K2 = sys_par[3]
    
    K1=K1/M
    K2=K2/M
    
    
    diag_x_x  = -2*K1*np.sin(qx)-K2*(np.sin(qx+qy)+np.sin(qx-qy))
    diag_y_x  = -K2*(np.sin(qx+qy)+np.sin(qx-qy))
    off_d_x = -K2*(np.sin(qx+qy)-np.sin(qx-qy))


    
    diag_x_y  = -K2*(np.sin(qy+qx)+np.sin(qy-qx))
    diag_y_y  = -2*K1*np.sin(qy)-K2*(np.sin(qy+qx)+np.sin(qy-qx))
    off_d_y = -K2*(np.sin(qy+qx)-np.sin(qy-qx))


    
    ax = vec[0]*vec[0]*diag_x_x + vec[5]*vec[5]*diag_y_x + 2*(vec[0]*vec[5])*off_d_x  #expectation value of the derivative operator   
    ay = vec[0]*vec[0]*diag_x_y + vec[5]*vec[5]*diag_y_y + 2*(vec[0]*vec[5])*off_d_y  #expectation value of the derivative operator
    vx = -ax/(2*np.sqrt(w+0.000000001))
    vy = -ay/(2*np.sqrt(w+0.000000001))
    
    return np.array([vx,vy])


    
def Eig(qx,qy,sys_par):
    M_intra = sys_par[0]
    M_inter = Matrix_inter(qx,qy,sys_par)
    eigVal, eigVec = np.linalg.eigh(M_intra+M_inter)
    return -eigVal, eigVec.transpose()



def rescale_K(M,m,K):
    factors = np.array([np.sqrt(M*m[0]),np.sqrt(M*m[1]),np.sqrt(M*m[2]),np.sqrt(M*m[3]),np.sqrt(m[0]*m[1]),np.sqrt(m[1]*m[2]),np.sqrt(m[2]*m[3]),np.sqrt(m[3]*m[0]),np.sqrt(m[0]*m[2]),np.sqrt(m[1]*m[3])])
    return K/factors


def Matrix_inter(qx,qy,sys_par):
    
    M  = sys_par[1]
    K1 = sys_par[2]
    K2 = sys_par[3]
    
    K1=K1/M
    K2=K2/M
    
    diag_x  = -4*K1*(np.sin(qx/2)**2)-2*K2*(np.sin((qx+qy)/2)**2+np.sin((qx-qy)/2)**2)
    diag_y  = -4*K1*(np.sin(qy/2)**2)-2*K2*(np.sin((qx+qy)/2)**2+np.sin((qx-qy)/2)**2)
    off_d = -2*K2*(np.sin((qx+qy)/2)**2-np.sin((qx-qy)/2)**2)
    
    res = np.zeros([10,10])
    
    res[0][0] = diag_x
    res[5][5] = diag_y
    res[0][5] = off_d
    res[5][0] = off_d
    
    return res
    
    

def Matrix_intra(M,m,K_intra):
    
    K = rescale_K(M,m,K_intra)
    
    res = np.zeros([10,10])
    #spring 1
    res[5][5]-=K[0]
    res[5][6]+=K[0]
    res[6][6]-=K[0]
    res[6][5]+=K[0]
    #spring 2
    res[0][0]-=K[1]
    res[0][2]+=K[1]
    res[2][2]-=K[1]
    res[2][0]+=K[1]
    #spring 3
    res[5][5]-=K[2]
    res[5][8]+=K[2]
    res[8][8]-=K[2]
    res[8][5]+=K[2]
    #spring 4
    res[0][0]-=K[3]
    res[0][4]+=K[3]
    res[4][4]-=K[3]
    res[4][0]+=K[3]
    #spring 5
    res[1][1]-=K[4]/2
    res[1][2]+=K[4]/2
    res[1][6]+=K[4]/2
    res[1][7]-=K[4]/2
    res[6][1]+=K[4]/2
    res[6][2]-=K[4]/2
    res[6][6]-=K[4]/2
    res[6][7]+=K[4]/2

    res[2][2]-=K[4]/2
    res[2][1]+=K[4]/2
    res[2][7]+=K[4]/2
    res[2][6]-=K[4]/2
    res[7][2]+=K[4]/2
    res[7][1]-=K[4]/2
    res[7][7]-=K[4]/2
    res[7][6]+=K[4]/2

    #spring 6
    res[2][2]-=K[5]/2
    res[2][3]+=K[5]/2
    res[2][7]-=K[5]/2
    res[2][8]+=K[5]/2
    res[7][2]-=K[5]/2
    res[7][3]+=K[5]/2
    res[7][7]-=K[5]/2
    res[7][8]+=K[5]/2

    res[3][3]-=K[5]/2
    res[3][2]+=K[5]/2
    res[3][8]-=K[5]/2
    res[3][7]+=K[5]/2
    res[8][3]-=K[5]/2
    res[8][2]+=K[5]/2
    res[8][8]-=K[5]/2
    res[8][7]+=K[5]/2

    #spring 7
    res[3][3]-=K[6]/2
    res[3][4]+=K[6]/2
    res[3][8]+=K[6]/2
    res[3][9]-=K[6]/2
    res[8][3]+=K[6]/2
    res[8][4]-=K[6]/2
    res[8][8]-=K[6]/2
    res[8][9]+=K[6]/2

    res[4][4]-=K[6]/2
    res[4][3]+=K[6]/2
    res[4][9]+=K[6]/2
    res[4][8]-=K[6]/2
    res[9][4]+=K[6]/2
    res[9][3]-=K[6]/2
    res[9][9]-=K[6]/2
    res[9][8]+=K[6]/2
    
    #spring 8
    res[4][4]-=K[7]/2
    res[4][1]+=K[7]/2
    res[4][9]-=K[7]/2
    res[4][6]+=K[7]/2
    res[9][4]-=K[7]/2
    res[9][1]+=K[7]/2
    res[9][9]-=K[7]/2
    res[9][6]+=K[7]/2

    res[1][1]-=K[7]/2
    res[1][4]+=K[7]/2
    res[1][6]-=K[7]/2
    res[1][9]+=K[7]/2
    res[6][1]-=K[7]/2
    res[6][4]+=K[7]/2
    res[6][6]-=K[7]/2
    res[6][9]+=K[7]/2
    
    #spring 9
    res[6][6]-=K[8]
    res[6][8]+=K[8]
    res[8][8]-=K[8]
    res[8][6]+=K[8]
    
    #spring 10
    res[2][2]-=K[9]
    res[2][4]+=K[9]
    res[4][4]-=K[9]
    res[4][2]+=K[9]
    
    return res

```

```python
def velocity_distribution(sys_par,q_mesh,vM,v_mesh,bdning):
    
    vx_ls = np.linspace(0,vM,v_mesh)
    vy_ls = np.linspace(0,vM,v_mesh)
    
    result = np.zeros([v_mesh,v_mesh])
    
    qx_ls = np.linspace(-np.pi,np.pi,q_mesh)
    qy_ls = np.linspace(-np.pi,np.pi,q_mesh)
    
    v_ls = []

    for qx in qx_ls:
        for qy in qy_ls:
            
            Val, Vec = Eig(qx,qy,sys_par)
            v_ls+= [phonon_v(Vec[i],np.sqrt(abs(Val[i])),qx,qy,sys_par) for i in range(9)]

    for i in range(len(vx_ls)):
        for j in range(len(vy_ls)):
            for v in v_ls:
                dV2= (vx_ls[i]-v[0])**2+(vy_ls[j]-v[1])**2

                result[i][j] += bdning/(dV2+bdning**2)
    
    return result
```

```python

sigma = 1
x = 8
def f1(vx,vy,w0 = 1):

    if vx > 0:
        return np.sqrt(vx**2+vy**2)/vx*np.exp(-w0**2*sigma**2/2/vx**2-x**2*vy**2/(vx**2+vy**2)/2/sigma**2)
    else:
        return 0
vc = 1
def f_linear(vx,w0 = 1):
    
    if vc <= vx:
        return 0    
    else:
        vy = np.sqrt(vc**2-vx**2)
        return (vc/vy)*(f1(vx,vy,w0)+f1(vx,-vy,w0)) 

def f_cos(V_MAX = 1.2, V_MESH = 1000 ,Q_MESH = 100, bdning = 3,w0 = 1):

    f_ls = np.zeros(V_MESH)

    for qx in np.linspace(-np.pi,np.pi,Q_MESH):
        for qy in np.linspace(-np.pi,np.pi,Q_MESH):
            vx = np.sin(qx)*np.cos(qy)
            vy = np.sin(qy)*np.cos(qx)
            mid = int(vx/V_MAX * V_MESH)
            temp = f1(vx,vy,w0)

            for i in range(max(mid-10*bdning,0),min(mid+10*bdning,V_MESH)):

                f_ls[i] = f_ls[i] + temp/np.sqrt(2*np.pi)/bdning * np.exp(-(((i-mid)/V_MESH * V_MAX-vx)**2)/2/bdning**2)/Q_MESH**2
    v_ls = np.linspace(0.0,V_MAX,V_MESH)
    
    return (v_ls,f_ls)

```

```python
Q_MESH = 100
V_MAX = 5
V_MESH = 1000
bdning = 5
def f_model():

    
    v_ls = np.linspace(0.0,V_MAX,V_MESH)


    for qx in np.linspace(-np.pi, np.pi, Q_MESH):
        for qy in np.linspace(-np.pi, np.pi, Q_MESH):
            val, vec = Eig(qx,qy,sys_par)
            for i in range(9):
                vx,vy = phonon_v(vec[i],abs(val[i])**0.5,qx,qy,sys_par)
                mid = int(vx/V_MAX * V_MESH)
                temp = f1(vx,vy)                
                for i in range(max(mid-3*bdning,0),mid+3*bdning):
                    f_ls[i] += temp/np.sqrt(2*np.pi)/bdning * np.exp(-(((i-mid)/V_MESH * V_MAX-vx)**2)/2/bdning**2)/Q_MESH**2
    return v_ls,f_ls


```

```python
v_ls,f_ls = f_model()
```
